```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image

def show_img(results, title):
    for r in results:
        im_array = r.plot()
        im = Image.fromarray(im_array[..., ::-1])
        plt.imshow(im)
        plt.title(title)
        plt.axis('off')
        plt.show()

model = YOLO('yolov8n.pt')

# Perform prediction
results = model('https://ultralytics.com/images/bus.jpg')

# Display the results
show_img(results, 'YOLOv8')
```

```python
def show_img_grid(results, ax, title):
    for r in results:
        im_array = r.plot()
        im = Image.fromarray(im_array[..., ::-1])
        ax.imshow(im)
        ax.set_title(title)
        ax.axis('off')

# Load the models
models = {
    'YOLOv8': YOLO('yolov8n.pt'),
    'YOLOv9': YOLO('yolov9t.pt'),
    'YOLOv10': YOLO('yolov10n.pt'),
    'YOLOv11': YOLO('yolo11n.pt')
}

# Perform predictions for each model
results = {}
for name, model in models.items():
    results[name] = model('https://ultralytics.com/images/bus.jpg')

# Display results in a 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
axs = axs.flatten()

for i, (name, result) in enumerate(results.items()):
    show_img_grid(result, axs[i], name)  # Pass the axis object to show_img

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

## Object Tracking with YOLO v11

```python
from pytubefix import YouTube

video_path = YouTube('https://www.youtube.com/watch?v=Ina7KMV2OEI').streams.first().download()
```

```python
import cv2

# Load the model
model = YOLO('yolo11n.pt')

# Open the video capture
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection with tracking
    results = model.track(frame, persist=True, verbose=False)

    # Check if results are not None before plotting
    if results is not None and results[0] is not None:
        annotated_frame = results[0].plot()
        # Display the resulting frame
        cv2.imshow("Frame", annotated_frame)
    else:
        # Handle cases where no tracks are found
        cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import math
import time
from ultralytics import YOLO

# Dictionary to store YOLO model names and their corresponding model files
yolo_models = {
    "YOLOv8": "yolov8n.pt",
    "YOLOv9": "yolov9t.pt",
    "YOLOv10": "yolov10n.pt",
    "YOLOv11": "yolo11n.pt"
}

# Path to save the output video
output_path = 'output.mp4'

# Open the input video
cap = cv2.VideoCapture(video_path)

# Get video properties for the output video
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Calculate the number of frames for each model
frames_per_model = math.ceil(total_frames / len(yolo_models))

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Dictionary to store processing times for each model
model_processing_times = {}

# Iterate through YOLO versions
model_index = 0
for model_name, model_file in yolo_models.items():
    print(f"Processing with {model_name}...")

    # Load the YOLO model
    try:
        model = YOLO(model_file)
    except Exception as e:
        print(f"Error loading {model_name}: {e}")
        continue

    # Seek to the start frame for the current model
    start_frame = model_index * frames_per_model
    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    # Time the processing for the current model
    start_time = time.time()

    # Reset frame_count for each model
    model_frame_count = 0

    while model_frame_count < frames_per_model:
        ret, frame = cap.read()
        if not ret:
            break

        # Perform object detection with the current model
        results = model.track(frame, persist=True, verbose=False)

        # Get the annotated frame
        if results:
            annotated_frame = results[0].plot()

            # Add title and time to the frame
            title_text = f"Model: {model_name}"
            time_text = ""
            if model_name in model_processing_times:
                avg_time = model_processing_times[model_name] / model_frame_count  # Correct calculation
                time_text = f"Avg. Time: {avg_time:.3f}s/frame"

            cv2.putText(annotated_frame, title_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2, cv2.LINE_AA)
            cv2.putText(annotated_frame, time_text, (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2, cv2.LINE_AA)

            # Write the annotated frame to the output video
            out.write(annotated_frame)
        else:
            out.write(frame)

        # Optional: Display the frame
        # cv2.imshow(f"{model_name} Output", annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        model_frame_count += 1

    # Calculate and store the processing time
    end_time = time.time()
    model_processing_times[model_name] = end_time - start_time  # Store total time

    model_index += 1

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video saved to: {output_path}")
```

```python
import cv2
import time
from ultralytics import YOLO
import numpy as np

# Dictionary to store YOLO model names and their corresponding model files
yolo_models = {
    "YOLOv8": "yolov8n.pt",
    "YOLOv9": "yolov9t.pt",
    "YOLOv10": "yolov10n.pt",
    "YOLOv11": "yolo11n.pt"
}

# Path to save the output video
output_path = 'output.mp4'

# Open the input video
cap = cv2.VideoCapture(video_path)

# Get video properties for the output video
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Number of models
num_models = len(yolo_models)

# Calculate grid dimensions (2x2 in this case)
grid_rows = 2
grid_cols = 2

# Output video dimensions (double the width and height for the grid)
output_width = frame_width * grid_cols
output_height = frame_height * grid_rows

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4
out = cv2.VideoWriter(output_path, fourcc, fps, (output_width, output_height))

# Dictionary to store processing times and frames for each model
model_data = {}
for model_name in yolo_models.keys():
    model_data[model_name] = {"processing_time": 0, "frames": []}

# Iterate through YOLO versions
for model_name, model_file in yolo_models.items():
    print(f"Processing with {model_name}...")

    # Load the YOLO model
    try:
        model = YOLO(model_file)
    except Exception as e:
        print(f"Error loading {model_name}: {e}")
        continue

    # Reset video capture to the beginning
    cap.set(cv2.CAP_PROP_POS_FRAMES, 0)

    # Time the processing for the current model
    start_time = time.time()

    # Process all frames for the current model
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Perform object detection with the current model
        results = model.track(frame, persist=True, verbose=False)

        # Get the annotated frame
        if results:
            annotated_frame = results[0].plot()

            # Add title to the frame
            title_text = f"Model: {model_name}"
            cv2.putText(
                annotated_frame,
                title_text,
                (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0, 255, 0),
                2,
                cv2.LINE_AA,
            )

            model_data[model_name]["frames"].append(annotated_frame)
        else:
            model_data[model_name]["frames"].append(frame)  # Append original frame if no detection

        frame_count += 1

    # Calculate and store the processing time
    end_time = time.time()
    model_data[model_name]["processing_time"] = end_time - start_time

# Create a blank canvas for the grid
grid_frame = np.zeros((output_height, output_width, 3), dtype=np.uint8)

# Write frames to the output video, arranging them in a grid
frame_index = 0
while frame_index < total_frames:
    for i in range(grid_rows):
        for j in range(grid_cols):
            model_idx = (i * grid_cols + j) % num_models
            model_list = list(yolo_models.keys())
            current_model = model_list[model_idx]

            if frame_index < len(model_data[current_model]["frames"]):
                frame = model_data[current_model]["frames"][frame_index]

                # Add time to the frame
                time_text = ""
                if model_data[current_model]["processing_time"] > 0:  # Check for non-zero time
                    avg_time = model_data[current_model]["processing_time"] / len(
                        model_data[current_model]["frames"]
                    )
                    time_text = f"Avg. Time: {avg_time:.3f}s/frame"

                cv2.putText(
                    frame,
                    time_text,
                    (10, 60),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    0.8,
                    (0, 255, 0),
                    2,
                    cv2.LINE_AA,
                )

                # Place the frame in the correct grid position
                grid_frame[
                    i * frame_height : (i + 1) * frame_height,
                    j * frame_width : (j + 1) * frame_width,
                ] = frame

    # Write the grid frame to the output video
    out.write(grid_frame)
    frame_index += 1

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video saved to: {output_path}")
```
